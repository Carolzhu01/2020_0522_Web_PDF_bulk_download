# Bulk Download MFC financial report in pdf

```python
from bs4 import BeautifulSoup as bs
import requests as rq
import os

mydomain = 'http://manulife.com'
myurl = 'https://www.manulife.com/en/investors/results-and-reports.html'
filetype = '.pdf'

def get_soup(url):
    return bs(rq.get(url).text, 'html.parser')
```

### Create download function

```python
def download_file(url):
    local_filename = url.split('/')[-1]
    
    with rq.get(url) as r:
        with open(local_filename, 'wb') as f:
            f.write(r.content)
    print(local_filename)
    return local_filename
```

### Loop the web to download pdf file

```python
soup = get_soup(myurl)
for content in soup.find_all('a'):
    try:
        all_link = content['href']
    except:
        pass
    new_link = mydomain + all_link
    if new_link.endswith(filetype):
        try:
            download_file(new_link)
        except:
            pass
```
